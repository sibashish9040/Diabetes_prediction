## Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn import svm
from sklearn.metrics import accuracy_score
```

## Data collection and analysis

```python
dataset = pd.read_csv('diabetes.csv')
```

```python
# Printing the 1st 5 rows of the data
dataset.head()
```

```python
# number of rows and columns in data
dataset.shape
```

```python
# Statistical measures of the dataframe
dataset.describe()
```

```python
dataset['Outcome'].value_counts()
```

Here the labels


*   0 ---> Non- diabetic
*   1 ---> Diabetic



```python
# Lets see parameters difference between diabetic and non diabetic persons
dataset.groupby('Outcome').mean()
```

```python
# separating data and variables
x = dataset.drop(columns = 'Outcome', axis = 1)
y = dataset.Outcome
```

```python
print(x)
print(y)
```

# Data standardization
As there are different parameters and all of the ranges of them are very different form each other so we perform standarization in a particular range

```python
scaler = StandardScaler()
scaler.fit(x)
std_data = scaler.transform(x)
print(std_data)
```

```python
x = std_data
```

## Train test splitting
* test_size indicates how much of data we want to be test data i.e if it is 0.2 that means wee want 20% of the datas to be test and 880% for training
* stratify means it will split y into equal proportion . because it might happen that all diabetic cases fall into training and non diabetic into test ,so it may affect the model
* random state is used to reproduce the same type of split for future evaluation process


```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, stratify = y, random_state = 2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

## Training of model

```python
classifier = svm.SVC(kernel = 'linear')
```

```python
# training the support vector machine classifier
classifier.fit(x_train, y_train)
```

## Model evaluation
finding the accuracy score

```python
x_train_accuracy = classifier.predict(x_train)
training_data_accuracy = accuracy_score(x_train_accuracy, y_train)
print("The accuracy score for trainign data is:", training_data_accuracy)
```

```python
# accuracy score on test data
x_test_prediction = classifier.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
print("The accuracy score for test data is:", test_data_accuracy)
```

# Makign a predictive system

```python
input_data = (4, 110, 92, 0, 0, 37.6, 0.191, 30)
# converting input data to numpay array
input_data_as_numpy_array = np.asarray(input_data)
# reshaping the array as we are predicting for one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)
# as we have stadndardize the training data so we have to std it also
std_data = scaler.transform(input_data_reshaped)
print(std_data)

prediction = classifier.predict(std_data)
print(prediction)

if (prediction[0] == 0):
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
